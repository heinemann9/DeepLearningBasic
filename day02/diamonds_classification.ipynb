```python
from sklearn import tree
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('diamonds')
```

```python
X_data = df[['carat', 'depth', 'table']].values
y_data = df['price'].values
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2) # train과 test를 8:2로 분할
```

```python
model = tree.DecisionTreeClassifier(max_depth=3)
```

```python
model.fit(X_train, y_train)
```

```python
y = model.predict(X_train)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_train[:,0], X_train[:,1], X_train[:,2], c = y_train)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_train[:,0], X_train[:,1], X_train[:,2], c = y)
plt.show()
```

```python
print(model.score(X_train, y_train))
```

```python
print(model.score(X_test, y_test))
```

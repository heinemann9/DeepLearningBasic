```python
print(df.carat.unique())
print(df['cut'].unique())
print(df['color'].unique())
print(df['clarity'].unique())
print(df.depth.unique())
print(df.table.unique())
print(df.price.unique())
print(df.x.unique())
print(df.y.unique())
print(df.z.unique())
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

dfi = df.copy()

dfi['carat'] = label_encoder.fit_transform(dfi['carat'])
print(label_encoder.classes_)
dfi['cut'] = label_encoder.fit_transform(dfi['cut'])
print(label_encoder.classes_)
dfi['color'] = label_encoder.fit_transform(dfi['color'])
print(label_encoder.classes_)
dfi['clarity'] = label_encoder.fit_transform(dfi['clarity'])
print(label_encoder.classes_)
dfi['depth'] = label_encoder.fit_transform(dfi['depth'])
print(label_encoder.classes_)
dfi['table'] = label_encoder.fit_transform(dfi['table'])
print(label_encoder.classes_)
dfi['price'] = label_encoder.fit_transform(dfi['price'])
print(label_encoder.classes_)
dfi['x'] = label_encoder.fit_transform(dfi['x'])
print(label_encoder.classes_)
dfi['y'] = label_encoder.fit_transform(dfi['y'])
print(label_encoder.classes_)
dfi['z'] = label_encoder.fit_transform(dfi['z'])
print(label_encoder.classes_)
```

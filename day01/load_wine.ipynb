# load_wine

```python
import sklearn.datasets
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = sklearn.datasets.load_wine()
print(type(data))
print(data.keys())
```

```python
for f in data.keys():
    t = type(data[f])
    print('key: {}, type: {}'.format(f, t))
    if t == np.ndarray:
        print('shape: {}'.format(data[f].shape))
```

```python
print(data.feature_names)
```

```python
print(data.target_names)
```

```python
print(data.data)
```

```python
print(data.DESCR)
```

```python
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df['target'] = data.target
df['target'] = df['target'].apply(lambda x: data['target_names'][x])
```

```python
df.head()
```

```python
df.tail()
```

```python
print(df.shape)
```

```python
df.describe()
```

```python
df.plot()
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="target")
plt.show()
```

```python
sns.distplot(df[df.target == "class_0"]["alcohol"], hist=True, rug=True, label="class_0") #  , 
sns.distplot(df[df.target != "class_0"]["alcohol"], hist=True, rug=True, label="others")
plt.legend()
plt.show()
```

```python
sns.distplot(df[df.target == "class_0"]["alcohol"], hist=False, rug=True, label="class_0")
plt.legend()
plt.show()
```

```python
sns.FacetGrid(df,hue='target') \
.map(plt.scatter,'alcohol','proline') \
.add_legend()
```

```python
sns.FacetGrid(df,hue='target') \
.map(plt.scatter,'color_intensity','flavanoids') \
.add_legend()
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.violinplot(x='target', y='alcohol',data=df)
plt.subplot(2,2,2)
sns.violinplot(x='target', y='color_intensity',data=df)
plt.subplot(2,2,3)
sns.violinplot(x='target', y='flavanoids',data=df)
plt.subplot(2,2,4)
sns.violinplot(x='target', y='proline',data=df)
```

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(x='target', y='alcohol',data=df) #     , outlier 
plt.subplot(2,2,2)
sns.boxplot(x='target', y='flavanoids',data=df)
plt.subplot(2,2,3)
sns.boxplot(x='target', y='color_intensity',data=df)
plt.subplot(2,2,4)
sns.boxplot(x='target', y='proline',data=df)
```
